Sedaj dodamo excel funkcijo, ki nam izračuna št. pozitivnih primerov v zadnjih 7 dni in pa zadnjo vrednost hospitaliziranih in pa v kateri fazi se nahajamo:

````python
sheet["B1"] = "=SUM(B5:B11)/7"
sheet["B2"] = "=C5"

positive_7d_avg = sum([cell[0].value for cell in sheet["B5:B11"]])/7
hospitalized = sheet["C5"].value

if positive_7d_avg < 300:
    phase = "Zelena"
elif positive_7d_avg < 600 and hospitalized < 500:
    phase = "Rumena"
elif positive_7d_avg < 1000 and hospitalized < 1000:
    phase = "Oranžna"
elif positive_7d_avg < 1350 and hospitalized < 1250:
    phase = "Rdeča"
elif positive_7d_avg > 1350 or hospitalized > 1200:
    phase = "Črna"
    
sheet["B3"] = phase


workbook.save(filename="nas_excel.xlsx")
````

## Adding values to dataframe

Dodajmo še 2 stolpca:
* 7day_avg , v katerem imamo vrednost povprečno vrednost pozitivnih zadnjih 7 dni
* phase - v kateri imamo string v kateri fazi bi bili

````python
for index, row in data.iterrows():
    #print(index, " neki ", row)
    if index+6 >= data.shape[0]:
        break
    positive_7d_avg = sum(data.loc[index:index+6, "Positive"]) / 7
    data.loc[index, "7d_avg_positive"] = positive_7d_avg
    
    phase = ""
    if positive_7d_avg < 300:
        phase = "zelena"
    elif positive_7d_avg < 600 and row["Hospitalized"] < 500:
        phase = "rumena"
    elif positive_7d_avg < 1000 and row["Hospitalized"] < 1000:
        phase = "oranžna"
    elif positive_7d_avg < 1350 and row["Hospitalized"] < 1200:
        phase = "rdeča"
    elif positive_7d_avg > 1350 and row["Hospitalized"] > 1200:
        phase = "črna"
    data.loc[index, "phase"] = phase
    
    
data.head(30)
        
````

# Simple plot

Začeli bomo s preprostim primerom:
> **izrisali bomo funkcijo sinus in kosinus na isti graf**.

Za začetek bomo uporabili raznorazne default vrednosti, za velikost grafa, barvo črt, stil črt, itd.., nato pa bomo graf nadgrajevali.

````python
#x-os
X = np.linspace(-np.pi, np.pi, 256) # dobimo 256 vrednosti od -pi do vključno pi

#y-os, potrebujemo vrednsoti Cos(x) in sin(x)
C = np.cos(X)
S = np.sin(X)

plt.plot(X,C)
plt.plot(X,S)

plt.show()
````

# Kako Matplotlib izrisuje stvari (Figures, Subplots, Axes, Ticks)

Matplotplib nam bo po defaultu ustvaril `figure`. Po defaultu bo znotraj `figure` ustvaril 1 `subplot`, ki bo zavzel celotno `figuro` in znotraj `subplota` nam bo izrisal graf.

 **Figure** v matplotlib-u pomeni celotno okno. Za naslov imajo `Figure _cifra_`, številčenje pa se začne z 1.
 Ostali parametri, ki definirajo `figure` so:
 * figsize - velikost figure v inčih (dolžina, višina)
 * dpi - resolucija v dots per inch
 * facecolor - barva risalne podlage
 * edgecolor - barva obrobe
 
**Subplot**

Z uporabo `subplot` lahko našo figuro razdelamo na več razdelkov. Specificiramo število vrstic, število stolpcev in številko `plot-a`.

````python
# Create a new figure of size 8x6 points, using 100 dots per inch
plt.figure(figsize=(8,6))

X = np.linspace(-np.pi, np.pi, 256)
C,S = np.cos(X), np.sin(X)

# Create a new subplot from a grid of 1x1
a = plt.subplot(2,2,2)
a.plot(X, C)
a.plot(X, S)


b = plt.subplot(2,2,3)
b.plot(X, C)
b.plot(X, S)


# Show result on screen
plt.show()
````

**Axes** se obnašajo podobno kot `subplot`, le da oni lahko ležijo kjerkoli v figuri.

`axes([left, bottom, width, height])`

````python
# Create a new figure of size 8x6 points, using 100 dots per inch
plt.figure(figsize=(6,6), dpi=100)

X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

# Create a new subplot from a grid of 1x1
plt.axes([0.1, 0.1, 0.8, 0.8])
plt.plot(X, C)
plt.plot(X, S)


plt.axes([0.2, 0.2, 0.3, 0.3])
plt.plot(X, C)
plt.plot(X, S)

# Show result on screen
plt.show()
````

### Spreminjanje lastnosti okvira

> Spines so linije na katerih se nahajajo naši "tick marks" in trenutno omejujejo naš graf. Lahko so postavljene na poljubna mest. Za naš primer bomo 2 postavili na sredino preostali dve pa izbrisali tako, da jima ne bomo dodelili barve.

> Pozicija spine je definirana s tuplom dveh vrednosti (position type, amount). Tipi pozicije so:
* 'outward' : postavi spine zunaj našega "data area" za specificirano število točk. Negativno število postavi spine znotraj "data area"

## Animation

`FuncAnimation(fig, update, interval=10)`
* fig - figura katero animiramo
* update - funkcija katero se kliče vsak nov frame. Frame spremenljivka je vedno podana kot prva, ostalo so naši poljubni parametri
* interval - delay med frami v milisekindah. Default je 200

````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook

fig = plt.figure(figsize=(6,6), facecolor='white')

ax = plt.subplot(1,1,1)

X = np.linspace(-np.pi, np.pi, 256)
C,S = np.cos(X), np.sin(X)
plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-", label="cosine")
plt.plot(X, S, color="red", linewidth=5.0, linestyle="--",label="sine")

def update(frame):
    global X, ax, C, S
    X = np.delete(X, 0)
    X = np.append(X, X[-1]+(X[-1]-X[-2]))
    C,S = np.cos(X), np.sin(X)
    ax.clear()
    ax.plot(X, C, color="blue", linewidth=1.0, linestyle="-", label="cosine")
    ax.plot(X, S, color="red", linewidth=5.0, linestyle="--",label="sine")

animation = FuncAnimation(fig, update, interval = 40) # 40 milisekund je od oka 24FPS
plt.show()
````

## Bar plot

````python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib notebook

n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.axes([0.025,0.025,0.95,0.95])
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

for x,y in zip(X,Y2):
    plt.text(x+0.4, -y-0.05, '%.2f' % y, ha='center', va= 'top')

plt.xlim(-.5,n), plt.xticks([])
plt.ylim(-1.25,+1.25), plt.yticks([])

plt.show()
````

## Table

Ustvarjanje in populacijo tabele iz že znanih podatkov lahko dosežemo s preprostim **for** loopom.

````python
document.add_paragraph()

table_head = ("#", "ISBN", "Description")
podatki = [(1, "10: 0316029181", "The Last Wish: Introducing The Witcher"),
          (2, "10: 0316390682", "Leviathan Wakes"),
          (3, "10: 1408855658", "Harry Potter and the Philosopher's Stone")]


table2 = document.add_table(1, 3, style='Table Grid')

table_head_row = table2.rows[0].cells
table_head_row[0].text = table_head[0]
table_head_row[1].text = table_head[1]
table_head_row[2].text = table_head[2]

for podatek in podatki:
    cells = table2.add_row().cells
    cells[0].text = str(podatek[0])
    cells[1].text = podatek[1]
    cells[2].text = podatek[2]

document.save('nas_doc.docx')
````

## Styles

Paragraf sam po sebi drži informacijo kot je indentations, tabs, line height itd..

Bold, italic, etc. informacije pa so definirano znotraj **run level**. 1 paragraf ima lahko več **run levels**.

````python
from docx.shared import RGBColor

paragraph = document.add_paragraph('Lorem ipsum ')
run = paragraph.add_run('dolor')
run.bold = True
font = run.font
font.color.rgb = RGBColor(0xFF, 0x00, 0xFF)

paragraph.add_run(' sit amet.')

document.save('nas_doc.docx')
````

Run lahko dodamo tudi specifičen stil:

````python
paragraph = document.add_paragraph('Normal text, ')
paragraph.add_run('text with emphasis.', 'Emphasis')

document.save('nas_doc.docx')
````
