## Headers

## Querry string parameters

```python
url = "https://api.github.com/search/repositories"
response = requests.get(url, params={"q": "requests+language:python"})

data = response.json()
print(data)
```

```python
url = "https://api.github.com/search/repositories"
response = requests.get(url, params=[("q", "requests+language:python")])

data = response.json()
print(data)
```

## Requests headers

```python
url = "https://api.github.com/search/repositories"
response = requests.get(url, 
                        params={"q": "requests+language:python"},
                       headers={"Accept": "application/vnd.github.v3.text-match+json"})

data = response.json()
print(data)
```

# Other HTTP methods:

```python
requests.post("https://httpbin.org/post", data={"key":"value"})
requests.put("https://httpbin.org/put", data={"key":"value"})
requests.delete("https://httpbin.org/delete")
requests.head("https://httpbin.org/head")
requests.patch("https://httpbin.org/patch", data={"key":"value"})
requests.options("https://httpbin.org/get")
```

## Message body

```python
url = "https://httpbin.org/post"
data = {"podatek": "HelloWORLD!",
       1: 200}

response = requests.post(url, data=data)
print(response.text)
```

```python
url = "https://httpbin.org/post"
data = [("podatek", "HelloWORLD!"), (1, 200)]

response = requests.post(url, data=data)
print(response.text)
```

```python
url = "https://httpbin.org/post"
data = {"podatek": "HelloWORLD!",
       1: 200}

response = requests.post(url, json=data)
print(response.text)
```

## Inspecting request

```python
url = "https://httpbin.org/post"
data = {"podatek": "HelloWORLD!",
       1: 200}

response = requests.post(url, json=data, headers={"Content-Type": "application/json"})
print(response.request.headers)
print(response.request.body)
```

## Authentication

```python
url = "https://httpbin.org/basic-auth/admin/12345"

response = requests.get(url, auth=("admin", "12345"))
print(response)
print(response.json())
```

## Sessions

```python
response = requests.get("https://httpbin.org/cookies/set", params={"moj_cookie": "hello world!"})
print(response.text)
```

```python
response = requests.get("https://httpbin.org/cookies")
print(response.text)
```

```python
session = requests.Session()

response = session.get("https://httpbin.org/cookies/set", params={"moj_cookie": "hello world!"})
print(response.text)
```

```python
response = session.get("https://httpbin.org/cookies")
print(response.text)
```

```python
session = requests.Session()
session.auth = ("admin", "12345")

url = "https://httpbin.org/basic-auth/admin/12345"
response = session.get(url)
print(response.json())
```

# Realen primer 1

```python
url = "https://opendata.si/promet/bicikelj/list/"
response = requests.get(url)
data = response.json()

free_bike = False
free_park = False

for key, station in data["markers"].items():
    if station["name"] == "VILHARJEVA CESTA":
        print(station)
        print()
        if int(station["station"]["available"]) > 0:
            free_bike = True
        
    if station["name"] == "STARA CERKEV":
        print(station)
        print()
        if int(station["station"]["free"]) > 0:
            free_park = True
            
            
if free_bike and free_park:
    print("Lahko se odpelješ")
else:
    print("Ne morš s kolesom.")
```

```python
url = "https://api.coingecko.com/api/v3/simple/price"
params = {"ids": "bitcoin",
         "vs_currencies": "eur",
         "include_market_cap": "true",
         "include_24hr_vol": "false",
         "include_24hr_change": "false",
         "include_last_updated_at": "false"}

response = requests.get(url, params=params)
data = response.json()
print(data)
```
